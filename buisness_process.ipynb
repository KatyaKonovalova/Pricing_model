1. Пользователь загружает в систему csv-файл
2. Система сохраняет данные в БД Postgres
3. Система два раза в сутки выгружает все данные из БД и делает прогноз
4. Пользователь заходит в веб-интерфейс, где может посмотреть прогноз

**Пользователь загружает в систему csv-файл.**

1. Изначальный csv-файл, который дан, сами закидываем в БД.
2. Последующие файлы можно загружать через сайт, который будет принимать файл и передавать его в БД.
3. После получения нового файла, данные по ценам, проданному количеству товаров и затратам на производство должны заменяться на новые, если такой товар уже есть или добавляться, если это новый товар.

**Права**

1. Загружать файлы на сайт могут только пользователи, у которых для этого есть права.
2. После входа в свой аккаунт, если у пользователя есть права, то есть он является Дата инженером, у него будет кнопка загрузки ежедневного отчета.
3. Если пользователь является Аналитиком или не является Дата инженером, то у него будет отсутствовать кнопка загрузки отчета, и будут права только для запроса.

**Данные csv**

1. Файл загруженный на сайт, должен попадать в папку data
2. После этого должен отправляться в БД diploma
3. Надо сделать общий параметр и метод для передачи файла в БД

**БД**

1. БД должна обновляться после получения нового файла
2. Если продукт и компания, производящая его, уже есть в таблице и в новом файле они также есть, то происходит замена данных: цены, количества проданного продукта, затрат на производство.
3. Если этого продукта нет, то появляется новая строка.
4. При обновлении данных и занесении нового продукта, данные по дате также обновляются. Но только у новых или обновленных объектов.
5. Обновление происходит каждый день вечером (но это не столь важно)
